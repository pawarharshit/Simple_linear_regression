```python
#importing the library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing the dataset
dataset_train = pd.read_csv('train.csv')
dataset_test = pd.read_csv('test.csv')
print(dataset_train.shape)
print(dataset_train.head())
```

```python
#checking and removing null values
dataset_train = dataset_train.dropna()
dataset_test = dataset_test.dropna()
print(dataset_train.shape)
```

```python
#splitting the data into dependent and independent variable
X_train = dataset_train.iloc[:,:-1].values
y_train = dataset_train.iloc[:,1:2].values
print(np.shape(X_train),"  ",np.shape(y_train))
```

```python
#checking for skew data if mean ~ median then data is not skew
print(np.mean(X_train))
print(np.median(X_train))

print(np.mean(y_train))
print(np.median(y_train))
```

```python
#plotting actual data
print("Graph of actual data")
plt.scatter(X_train,y_train,s = 5,c = 'red')
plt.xlabel('X_train')
plt.ylabel('y_train')
plt.show()
```

```python
#importing the  linear model
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
#fit the model into our data
lm.fit(X_train,y_train)
```

```python
print(lm.score(X_train,y_train))
```

```python
y_pred_train = lm.predict(X_train)
```

```python
for i in range(5):
    print(y_train[i],"  ",y_pred_train[i])
```

```python
#plotting the predicting and actual training data
plt.scatter(X_train,y_train,s = 4,c='red')
plt.plot(X_train,lm.predict(X_train),c = 'blue')
plt.title("training performance")
plt.xlabel('X_train')
plt.ylabel('Y')
```

```python
#check model on test data
X_test = dataset_test.iloc[:,:-1].values
y_test = dataset_test.iloc[:,1:2].values
```

```python
y_pred_test = lm.predict(X_test)
```

```python
for i in range(5):
    print(y_test[i],"   ",y_pred_test[i])
```

```python
#Plotting the test data
plt.scatter(X_test,y_test,s = 5,c='red')
plt.plot(X_train,lm.predict(X_train),c = 'purple')
plt.title("training performance")
plt.xlabel('X_train')
plt.ylabel('Y')
```
